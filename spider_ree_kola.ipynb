```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spiders = pd.read_excel('Table Kola dykes Rock compositions.xlsx', sheet_name='spiders')
ree = pd.read_excel('Table Kola dykes Rock compositions.xlsx', sheet_name='РЗЭ')
```

```python
spiders
```

```python
exclude = ['Location', 'Rock']
spiders_filtered = spiders[-spiders['Sample'].isin(exclude)]
df = pd.DataFrame(spiders_filtered)
df['18КL-21'] = df['18КL-21'].astype('float')
```

```python
df['18КL-21']
```

```python
plt.figure(figsize=(15, 10))
sns.set(style="ticks", palette="colorblind", font="Arial")
spidergram = sns.lineplot(
                x="Sample",
                y="18КL-21",
                data=df,
                sort=False
                )
spidergram.set(yscale="log")
```
