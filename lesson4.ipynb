```python
stations = pd.read_csv(open('data-398-2019-12-31.csv', 'r', encoding='cp1251'), sep=';')
pavilion_count = stations.groupby(['Pavilion']).size()
pavilion_count
```

```python
plt.pie(pavilion_count, labels=pavilion_count.index, autopct='%.2f')
plt.show()
```

Уберем из столбца Pavilion значения "Pavilion", чтобы остались только "да" и "нет".

```python
exclude = ['Pavilion']
stations_filtered = stations[-stations['Pavilion'].isin(exclude)]
pavilion_count_filtered = stations_filtered.groupby(['Pavilion']).size()
```

```python
pavilion_count_filtered
plt.pie(pavilion_count_filtered, labels=pavilion_count_filtered.index, autopct='%.2f')
plt.title('Павильоны у остановок')
plt.legend()
plt.show()
```

```python
admarea_count = stations.groupby(['AdmArea']).size()
admarea_count
```

Убираем значение AdmArea 11

```python
exclude = ['AdmArea']
admarea_filtered = stations[-stations['AdmArea'].isin(exclude)]
admarea_count_filtered = admarea_filtered.groupby(['AdmArea']).size()
```

```python
admarea_count_filtered
```

```python
labels = ['ВАО', 'ЗАО', 'ЗелАО', 'НовАО', 'САО', 'СВАО', 'СЗАО', 'ТроАО', 'ЦАО', 'ЮВАО', 'ЮЗАО', 'ЮАО']
```

```python
cmap = get_cmap('tab20')
colors = [cmap(i) for i in range(len(labels))]
plt.bar(labels, admarea_count_filtered, color=colors)
plt.title('Остановки по округам')
plt.show()
```
