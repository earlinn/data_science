```python
import csv
```

```python
streets = {}
```

```python
%time
with open('data-398-2019-12-31.csv', 'r', encoding='cp1251') as bus_stops_file:
    reader = csv.DictReader(bus_stops_file, delimiter=';')
    for row in reader:
        if row['Street'] in streets:
            streets[row['Street']] += 1
        else:
            streets[row['Street']] = 1
```

```python
streets
```

```python
%time
sorted(streets.items(), key=lambda item: -item[1])
```
