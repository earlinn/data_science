```python
import pandas as pd
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
export_2017 = pd.read_excel('Russia-2017.xlsx', sheet_name='Partner')
export_2018 = pd.read_excel('Russia-2018.xlsx', sheet_name='Partner')
```

```python
export = pd.concat([export_2017, export_2018])
```

```python
export.sample(10)
```

```python
exclude = [' World', 'Europe & Central Asia', 'East Asia & Pacific', 'Middle East & North Africa', 'United States Minor Outlying I']
export_filtered = export[-export['Partner Name'].isin(exclude)]
```

```python
export_pivot = export_filtered.pivot_table(index=['Year', 'Partner Name'], values='Export (US$ Thousand)', aggfunc='sum')\
    .sort_values(['Year', 'Export (US$ Thousand)'], ascending=[True, False])
```

```python
export_pivot[export_pivot.index.get_level_values('Year') == 2017]
```

```python
export_pivot[export_pivot.index.get_level_values('Year') == 2018]
```
